```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
qb_df = pd.read_csv("2010_QB_stats.csv")
```

```python
qb_df.head()
```

```python
qb_col = qb_df["qb"].tolist()
```

```python
qb_col
```

```python
qb_list = []
for item in qb_col:
    qb = item.split(".")[0]
    qb = qb[:-1]
    qb_list.append(qb)
qb_list
```

```python
qb_df["qb"] = qb_list
```

```python
qb_df.head()
```

```python
qb_df.drop(['att','cmp', 'yds'
           ,'ypa','td','int'
           ,'lg','sack','loss'
           ,'game_points','home_away'
           ,'year'], axis=1, inplace=True)
qb_df.rename(columns={'qb': 'qb_name', 'rate': 'qb_rating'}, inplace=True)
```

```python
qb_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/qb_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
qb_df.to_sql(name='qb_rating', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from qb_rating', con=engine).head()
```
